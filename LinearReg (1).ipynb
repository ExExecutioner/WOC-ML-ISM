```python
def feature_scale(x):
  m,n=x.shape
  for i in range(n):
    x[:,i]=(x[:,i]-np.mean(x[:,i]))/np.std(x[:,i])
  return x

```

```python
def cost(w,b,x,y):
  costj=0
  m,n=x.shape
  for i in range(m):
    costj+=(np.dot(w,x[i])+b-y[i])**2
  costj=costj/(2*m)
  return costj
```

```python
def gradient(w,b,x,y):
  dw=np.zeros((n,))
  db=0

  for i in range(m):
    dd=np.dot(x[i],w)+b-y[i]
    for j in range(n):
      dw[j]+=dd*x[i,j]
    db+=dd
  dw=dw/m
  db=db/m
  return dw,db
```

```python
def gradient_decent(x,y,win,bin,alpha,iter,cost_func,grad_func,lambdaa):
  w=np.copy(win)
  b=bin
  cost_store=[]
  for i in range(iter):
    dw,db=gradient(w,b,x,y)
    w=w-alpha*dw
    b=b-alpha*db
    cost_store.append(cost_func(w,b,x,y))
  return w,b,cost_store
```
