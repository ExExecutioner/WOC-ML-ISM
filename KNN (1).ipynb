```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

train= np.array(pd.read_csv("multi_classification_train.csv"))
```

```python

print(train.shape)
```

```python
a,b=train.shape
```

```python
x_train=np.zeros([a,b-3])
y_train=np.zeros(a)
new=np.zeros(a)
for i in range(a):
  x_train[i]=train[i, 1:b-2]
  y_train[i]=train[i,b-2]
  new[i]=train[i,b-1]
```

```python
def k_NN(x,y,new):
  u , v = x.shape
  p = int(input("Enter the value of order(p): "))

  distance = np.zeros((u,2))
  for i in range(u):
    distance[i,1] = 0
    for j in range(v):
      distance[i,1] += (x[i,j] - new[j])**p
    distance[i,1] = np.power(distance[i,1],(1.0/p))
    distance[i,0] = i

  distance_asc = np.argsort(distance[:,1])
  distance = distance[distance_asc]

  k = int(input("Enter the value of k: "))
  K = np.zeros(k)
  for i in range(k):
    K[i]= y[int(distance[i,0])]

  unique , freq = np.unique(K,return_counts=True)
  freq_count = np.zeros((len(unique),2))
  for i in range(len(unique)):
    freq_count[i,0] = unique[i]
    freq_count[i,1] = freq[i]
  print(freq_count)
  class_predicted = freq_count[np.argmax(freq_count[:,1]),0]
  print(f"predicted class: {class_predicted}")
```

```python
k_NN(x_train,y_train,new)
```
