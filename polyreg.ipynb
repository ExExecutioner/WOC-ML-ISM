```python
def poly(k,x):
  m,n=x.shape
  xnew=np.empty([m,n*k])
  for i in range(m):
    count=0
    for j in range(n):
      for g in range(1,k+1):
        xnew[i][count]=(x[i][j])**g
        count+=1
  return xnew

```

```python
def feature_scale(x):
  m,n=x.shape
  for i in range(n):
    x[:,i]=(x[:,i]-np.mean(x[:,i]))/np.std(x[:,i])
  return x

```

```python
def cost(w,b,x,y):
  costj=0
  m,n=x.shape
  for i in range(m):
    costj+=(np.dot(w,x[i])+b-y[i])**2
  costj=costj/(2*m)
  return costj
```

```python
def gradient(w,b,x,y):
  m,n=x.shape
  dw=np.zeros((n,))
  db=0

  for i in range(m):
    dd=np.dot(x[i],w)+b-y[i]
    for j in range(n):
      dw[j]+=dd*x[i,j]
    db+=dd
  dw=dw/m
  db=db/m
  return dw,db
```

```python
def gradient_decent(x,y,win,bin,alpha,iter,cost_func,grad_func):
  w=np.copy(win)
  b=bin
  cost_store=[]
  for i in range(iter):
    dw,db=grad_func(w,b,x,y)
    w=w-alpha*dw
    b=b-alpha*db
    cost_store.append(cost_func(w,b,x,y))
  return w,b,cost_store
```
